```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Weight variations uniform
data = {"0,0.001": 0.928,
"1,1.001": 0.962,
"0,0.1": 0.931,
"0,10": 0.883,
"0,100": 0.77,
"50, 200": 0.72,
"500, 2000": 0.69,
"5000, 20000": 0.7}

fig = plt.figure()
plt.bar(range(len(data)), data.values())
plt.xticks(range(len(data)), data.keys(), rotation = "vertical")
plt.hlines(y=1, xmin=-0.5, xmax = len(data)- 0.5, color='r', linestyle='--', label='OPT')

plt.xlabel("Weight range (uniform)")
plt.ylabel("Accuracy")
plt.show()
```

```python
# BA training
data = {"3": 0.94,
        "5": 0.97,
        "7": 0.95
        }

fig = plt.figure()
plt.bar(range(len(data)), data.values())
plt.xticks(range(len(data)), data.keys(), rotation = "vertical")
plt.hlines(y=1, xmin=-0.5, xmax = len(data)- 0.5, color='r', linestyle='--', label='OPT')

plt.xlabel("BA parameter")
plt.ylabel("Accuracy")
plt.title("Generalization to BA")
plt.tight_layout()
plt.savefig("ba_generalization.png")
plt.show()
```

```python
# Generalization rideshare
data = {"32x32": 0.96,
        "50x50": 0.96,
        "100x100": 0.938,
        "250x250": 0.9
        }

fig = plt.figure()
plt.bar(range(len(data)), data.values())
plt.xticks(range(len(data)), data.keys(), rotation = "vertical")
plt.hlines(y=1, xmin=-0.5, xmax = len(data)- 0.5, color='r', linestyle='--', label='OPT')

plt.title("Generalization to Rideshare")
plt.xlabel("Graph size")
plt.ylabel("Accuracy")
plt.tight_layout()
plt.savefig('rideshare_generalization.png')
plt.show()
```
